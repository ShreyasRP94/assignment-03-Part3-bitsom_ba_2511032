# #Task 1 — File Read & Write Basics

```python
# Create a new text file and write initial content

# File name
file_name = "python_notes.txt"

# Writing initial content
with open(file_name, 'w', encoding='utf-8') as f:
    f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    f.write("Topic 2: Lists are ordered and mutable.\n")
    f.write("Topic 3: Dictionaries store key-value pairs.\n")
    f.write("Topic 4: Loops automate repetitive tasks.\n")
    f.write("\n")
    f.write("Topic 5: Exception handling prevents crashes.\n")

print("File written successfully.")
```

```python
#Appending more content to the existing file

with open(file_name, 'a', encoding='utf-8') as f:
    f.write("Topic 6: Functions help reuse code.\n")
    f.write("Topic 7: File handling allows persistent storage.\n")

print("Lines appended.")
```

```python
# Reading and displaying file contents

with open(file_name, 'r', encoding='utf-8') as f:
    lines = f.readlines()

print("File Contents:\n")

for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")
```

```python
# Counting total lines in the file

total_lines = len(lines)

print("\nTotal number of lines:", total_lines)
```

```python
# Searching for a keyword in the file

keyword = input("\nEnter keyword to search: ").strip().lower()

matches = []

for line in lines:
    if keyword in line.lower():
        matches.append(line.strip())

if matches:
    print("\nMatching lines:")
    for m in matches:
        print(m)
else:
    print("\nNo lines found containing the keyword.")
```

# #Task 2 — API Integration

```python
# API Interaction: Fetching data from a public API

import requests

url = "https://dummyjson.com/products?limit=20"

try:
    response = requests.get(url)
    response.raise_for_status()   # raises error if status != 200
    
    data = response.json()
    products = data['products']
    
    print("Products fetched successfully.\n")

except requests.exceptions.RequestException as e:
    print("Error fetching data:", e)
    products = []
```

```python
# Displaying product information in a formatted manner

print("ID  | Title                          | Category      | Price    | Rating")
print("----|--------------------------------|---------------|----------|-------")

for p in products:
    print(f"{p['id']:<3} | {p['title'][:30]:<30} | {p['category']:<13} | ${p['price']:<8} | {p['rating']}")
```

```python
# Filtering products with rating >= 4.5

filtered = []

for p in products:
    if p['rating'] >= 4.5:
        filtered.append(p)

# Sort by price descending
filtered.sort(key=lambda x: x['price'], reverse=True)

print("\nFiltered & Sorted Products (Rating ≥ 4.5):\n")

for p in filtered:
    print(f"{p['title']:<30} ${p['price']:<8} Rating: {p['rating']}")
```

```python
# Fetching laptops from the API

laptop_url = "https://dummyjson.com/products/category/laptops"

try:
    response = requests.get(laptop_url)
    response.raise_for_status()
    
    laptop_data = response.json()
    laptops = laptop_data['products']
    
    print("\nLaptops:\n")
    
    for item in laptops:
        print(f"{item['title']:<30} ${item['price']}")
        
except requests.exceptions.RequestException as e:
    print("Error fetching laptops:", e)
```

```python
# Creating a new product via POST request

post_url = "https://dummyjson.com/products/add"

payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

try:
    response = requests.post(post_url, json=payload)
    response.raise_for_status()
    
    result = response.json()
    
    print("\nPOST Response:\n")
    print(result)

except requests.exceptions.RequestException as e:
    print("Error in POST request:", e)
```

# #Task 3 — Exception Handling

## Part A - Guarded calculator

```python
# Exception handling example: Safe division function

def safe_divide(a, b):
    try:
        return a / b
    
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    
    except TypeError:
        return "Error: Invalid input types"
```

```python
#test cases for safe_divide function

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

## Part B — Guarded File Reader

```python
# Example of reading a file with exception handling

def read_file_safe(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            content = f.read()
            return content
    
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    
    finally:
        print("File operation attempt complete.")
```

```python
# Testing the read_file_safe function with a valid and an invalid file

print("\nReading valid file:\n")
print(read_file_safe("python_notes.txt"))

print("\nReading invalid file:\n")
print(read_file_safe("ghost_file.txt"))
```

```python
# API Interaction: Fetching data with error handling

import requests

url = "https://dummyjson.com/products?limit=20"

try:
    response = requests.get(url, timeout=5)
    response.raise_for_status()
    
    data = response.json()
    products = data['products']
    print("Data fetched successfully.")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

```python
# Creating a new product via POST request with error handling

post_url = "https://dummyjson.com/products/add"

payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

try:
    response = requests.post(post_url, json=payload, timeout=5)
    response.raise_for_status()
    
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

## Part D — Input Validation Loop

```python
while True:
    user_input = input("\nEnter a product ID (1–100) or 'quit': ").strip()
    
    if user_input.lower() == "quit":
        print("Exiting...")
        break
    
    # Validate integer
    if not user_input.isdigit():
        print("⚠ Please enter a valid number.")
        continue
    
    product_id = int(user_input)
    
    # Validate range
    if product_id < 1 or product_id > 100:
        print("⚠ ID must be between 1 and 100.")
        continue
    
    # API call
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)
        
        if response.status_code == 404:
            print("Product not found.")
        
        elif response.status_code == 200:
            product = response.json()
            print(f"Title: {product['title']} | Price: ${product['price']}")
        
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    
    except Exception as e:
        print("Error:", e)
```

# #Task 4 — Logging to File

```python
# Error logging function

from datetime import datetime

log_file = "error_log.txt"

def log_error(context, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    log_entry = f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n"
    
    with open(log_file, 'a', encoding='utf-8') as f:
        f.write(log_entry)
```

```python
# Simulating an API call that triggers an error to demonstrate logging

import requests

bad_url = "https://this-host-does-not-exist-xyz.com/api"

try:
    response = requests.get(bad_url, timeout=5)
    response.raise_for_status()

except requests.exceptions.ConnectionError as e:
    print("Connection error triggered.")
    log_error("fetch_products", "ConnectionError", str(e))

except requests.exceptions.Timeout as e:
    print("Timeout error triggered.")
    log_error("fetch_products", "Timeout", str(e))

except Exception as e:
    print("Unexpected error.")
    log_error("fetch_products", type(e).__name__, str(e))
```

```python
# Simulating an API call that triggers a 404 error to demonstrate logging

url = "https://dummyjson.com/products/999"

try:
    response = requests.get(url, timeout=5)
    
    if response.status_code != 200:
        print("HTTP error triggered.")
        
        log_error(
            "lookup_product",
            "HTTPError",
            f"{response.status_code} Not Found for product ID 999"
        )
    
    else:
        print("Product found (unexpected).")

except requests.exceptions.ConnectionError as e:
    log_error("lookup_product", "ConnectionError", str(e))

except requests.exceptions.Timeout as e:
    log_error("lookup_product", "Timeout", str(e))

except Exception as e:
    log_error("lookup_product", type(e).__name__, str(e))
```
