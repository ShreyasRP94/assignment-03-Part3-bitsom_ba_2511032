# #Task 1 — File Read & Write Basics

```python
# Create a new text file and write initial content

# File name
file_name = "python_notes.txt"

# Writing initial content
with open(file_name, 'w', encoding='utf-8') as f:
    f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    f.write("Topic 2: Lists are ordered and mutable.\n")
    f.write("Topic 3: Dictionaries store key-value pairs.\n")
    f.write("Topic 4: Loops automate repetitive tasks.\n")
    f.write("\n")
    f.write("Topic 5: Exception handling prevents crashes.\n")

print("File written successfully.")
```

```python
#Appending more content to the existing file

with open(file_name, 'a', encoding='utf-8') as f:
    f.write("Topic 6: Functions help reuse code.\n")
    f.write("Topic 7: File handling allows persistent storage.\n")

print("Lines appended.")
```

```python
# Reading and displaying file contents

with open(file_name, 'r', encoding='utf-8') as f:
    lines = f.readlines()

print("File Contents:\n")

for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")
```

```python
# Counting total lines in the file

total_lines = len(lines)

print("\nTotal number of lines:", total_lines)
```

```python
# Searching for a keyword in the file

keyword = input("\nEnter keyword to search: ").strip().lower()

matches = []

for line in lines:
    if keyword in line.lower():
        matches.append(line.strip())

if matches:
    print("\nMatching lines:")
    for m in matches:
        print(m)
else:
    print("\nNo lines found containing the keyword.")
```
